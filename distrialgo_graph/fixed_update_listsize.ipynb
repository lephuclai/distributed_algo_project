```python
import matplotlib.pyplot as plot
import csv
import numpy as np
import random
import glob
import scipy.stats

confidence_level = 0.95
z_value = scipy.stats.norm.ppf((1 + confidence_level) / 2)

num_threads = [1, 4, 6, 8, 10, 12]

coarse_grained = []
hands_over_hands = []
lazy_linked_list = []
```

```python
file = "data/fixed_update_listsize.csv"

with open(file, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if row[1] == "coarse_grained":
            coarse_grained.append(float(row[0]))
        if row[1] == "hands_over_hands":
            hands_over_hands.append(float(row[0]))
        if row[1] == "lazy_linked_list":
            lazy_linked_list.append(float(row[0]))        
            
print(coarse_grained)
print(hands_over_hands)
print(lazy_linked_list)
```

```python
tick_fontsize = 14
label_fontsize = 14
lg_fontsize = 10

plot.ylabel('Throughput (ops/s)', fontsize = label_fontsize)
plot.xlabel('Number of threads', fontsize = label_fontsize)

plot.plot(num_threads, coarse_grained, label = 'Coarse-grained')
plot.plot(num_threads, hands_over_hands, label = 'Hand-Over-Hand')
plot.plot(num_threads, lazy_linked_list, label = 'Lazy Linked List ')

plot.xticks(num_threads)

plot.legend(loc="best", fontsize = lg_fontsize)

plot.savefig('./pdf/fixed_update_listsize.pdf', bbox_inches='tight')

plot.show()
```
