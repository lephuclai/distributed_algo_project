```python
import matplotlib.pyplot as plot
import csv
import numpy as np
import random
import glob
import scipy.stats

confidence_level = 0.95
z_value = scipy.stats.norm.ppf((1 + confidence_level) / 2)

num_threads = [1, 4, 6, 8, 10, 12]

coarse_grained_100 = []
hands_over_hands_100 = []
lazy_linked_list_100 = []

coarse_grained_1000 = []
hands_over_hands_1000 = []
lazy_linked_list_1000 = []

coarse_grained_10000 = []
hands_over_hands_10000 = []
lazy_linked_list_10000 = []
```

```python
file = "data/fixed_update.csv"

with open(file, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if row[3] == '100':
            if row[1] == "coarse_grained":
                coarse_grained_100.append(float(row[0]))
            if row[1] == "hands_over_hands":
                hands_over_hands_100.append(float(row[0]))
            if row[1] == "lazy_linked_list":
                lazy_linked_list_100.append(float(row[0])) 

        if row[3] == '1000':
            if row[1] == "coarse_grained":
                coarse_grained_1000.append(float(row[0]))
            if row[1] == "hands_over_hands":
                hands_over_hands_1000.append(float(row[0]))
            if row[1] == "lazy_linked_list":
                lazy_linked_list_1000.append(float(row[0])) 

        if row[3] == '10000':
            if row[1] == "coarse_grained":
                coarse_grained_10000.append(float(row[0]))
            if row[1] == "hands_over_hands":
                hands_over_hands_10000.append(float(row[0]))
            if row[1] == "lazy_linked_list":
                lazy_linked_list_10000.append(float(row[0])) 

print(coarse_grained_100)
print(hands_over_hands_100)
print(lazy_linked_list_100)
print('--------------------------')

print(coarse_grained_1000)
print(hands_over_hands_1000)
print(lazy_linked_list_1000)
print('--------------------------')

print(coarse_grained_10000)
print(hands_over_hands_10000)
print(lazy_linked_list_10000)
print('--------------------------')
```

```python
tick_fontsize = 14
label_fontsize = 14
lg_fontsize = 10

plot.ylabel('Throughput (ops/s)', fontsize = label_fontsize)
plot.xlabel('Number of threads', fontsize = label_fontsize)

plot.plot(num_threads, coarse_grained_100, label = '100')
plot.plot(num_threads, coarse_grained_1000, label = '1000')
plot.plot(num_threads, coarse_grained_10000, label = '10000')

plot.xticks(num_threads)

plot.legend(loc="best", fontsize = lg_fontsize)

plot.savefig('./pdf/fixed_update_coarse_grained.pdf', bbox_inches='tight')

plot.show()
```
